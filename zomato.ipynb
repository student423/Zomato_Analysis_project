```python
#dlt unwanted data
df.dropna(inplace=True)
```

```python
# rename the colm name
df.rename(columns={'Currency':'Country'},inplace=True)
```

```python
# report count of country
df.Country.value_counts()
```

```python
df.columns
```

```python
# grouping by this colmn and which color is better or not
rating=df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
print(rating)
```

```python
# which color has how much record
rating=df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
print(rating)
```

```python
# i will change 0 to rating count
rating=df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'rating count'})
print(rating)
```

```python
## Observation
#1.when rating is b/w 4.5 to 4.9 ------> Excellent
#2.when rating is b/w 4.0 to 4.4 ------> Very good
#3.when rating is b/w 3.5 to 3.9 ------> Good
#4.when rating is b/w 2.5 to 3.4 ------> Average
#5.when rating is b/w 1.8 to 2.4 ------> Poor
#6.when rating is zero ------> not rating
```

```python
rating.head()
```

```python
## graph with seaborn
sns.barplot(x='Aggregate rating', y='rating count', data=rating)
sns.set(rc={'figure.figsize':(12,5)})
```

```python
# below the graph is much better than above graph and its clear everything what i did
sns.barplot(x='Aggregate rating', y='rating count', hue = 'Rating color', data=rating, palette=['White','Red','Orange','Yellow','Green','Green'])
sns.set(rc={'figure.figsize':(18,7)})
```

```python
## Observation
#1 Not rated count is very high
#2 Maximum number of rating are b/w 2.5 to 3.4
```

```python
sns.countplot(x='Rating color', data=rating, palette=['White','Red','Orange','Yellow','Green','Green'])
```
